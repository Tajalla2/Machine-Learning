```python
import pandas as pd
```

```python
df= pd.read_csv("bankloan.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop('ID',axis = 'columns')
df.head()
```

```python
import matplotlib.pyplot as plt
df.hist(figsize=(20,25),color='pink')
plt.show()
```

```python
data = df.drop('Personal.Loan',axis='columns')
data.head()
```

```python
target = df['Personal.Loan']
target
```

```python
final = pd.concat([data,target],axis=1)
```

```python
final.head()
```

```python
import seaborn as sns
plt.figure(figsize=(20,15))
ax = sns.heatmap(final.corr(),annot=True)
plt.show()
```

```python
sns.boxplot(x='CreditCard', y='Income',data=data)
plt.show()
```

```python
X = final.iloc[:,:-1]
y = final.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treemodel = DecisionTreeClassifier()
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(25,20))
tree.plot_tree(treemodel,filled=True)
```

```python
ypred = treemodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
score = accuracy_score(ypred,y_test)
print(score)
```

```python
print(classification_report(ypred,y_test))
```
